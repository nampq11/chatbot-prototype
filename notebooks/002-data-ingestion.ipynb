```python
def get_query_results(query, k=5):
    query_embedding = get_embedding(query)
    pipeline = [
        {
            "$vectorSearch": {
                "index": config.MONGO_COSMETIC_SURGEON_SEARCH_INDEX_NAME,
                "queryVector": query_embedding,
                "path": "embedding",
                "exact": True,
                "limit": k,
            }
        }, {
            "$project": {
                "_id": 0,
                "text": 1,
            }
        }
    ]
    results = collection.aggregate(pipeline)

    array_of_results = []
    for doc in results:
        array_of_results.append(doc)
    return array_of_results
```

```python
from langchain_mongodb.vectorstores import MongoDBAtlasVectorSearch

# Instantiate the vector store using your MongoDB connection string
embedding_model = get_embedding_model(
    model_id="BookingCare/gte-multilingual-base-v2.1",
    device="cpu"
)

vector_store = MongoDBAtlasVectorSearch.from_connection_string(
  connection_string = config.MONGO_URI,        # Atlas cluster or local deployment URI
  namespace = f"{config.MONGO_DB_NAME}.{config.MONGO_COSMETIC_SURGEON_COLLECTION}",  # Database and collection name
  embedding = embedding_model,                 # Embedding model to use
  index_name = "Cosmetic_surgeon_search_index",                      # Name of the vector search index
)
```

```python
from langchain_mongodb.retrievers.hybrid_search import MongoDBAtlasHybridSearchRetriever

retriever = MongoDBAtlasHybridSearchRetriever(
   vectorstore = vector_store,        # Vector store instance
   search_index_name = "search_index",  # Name of the Atlas Search index
   top_k = 5,                           # Number of documents to return
   fulltext_penalty = 60.0,             # Penalty for full-text search
   vector_penalty = 60.0                # Penalty for vector search
)

# Define your query
query = "bác sĩ Hoàng thanh tuấn"

# Print results
documents = retriever.invoke(query)
for doc in documents:
   print(doc)
```
