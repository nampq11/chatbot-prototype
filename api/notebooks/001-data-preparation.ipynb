```python
import pandas as pd
```

```python
import os
import glob
```

```python
data_path = '../data'
all_files = glob.glob(os.path.join(data_path, "*.csv"))
dfs = []

for file in all_files:
    df = pd.read_csv(file)
    print(file)
    print(df.columns)
    if "Việt Đức" in file:
        df = df.drop('Loại', axis=1)
    df['Bệnh viện'] = file.split('/')[-1].split('.')[0]
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True, axis=0)
combined_df.to_csv(os.path.join(data_path, "combined_data.csv"), index=False)
```

```python
from bs4 import BeautifulSoup
```

```python
def parse_html(text):
    try:
        return BeautifulSoup(text, 'html.parser').get_text()
    except:
        return text
```

```python
combined_df['Nội dung'] = combined_df['Nội dung'].apply(lambda x: parse_html(x))
combined_df
```

```python
combined_df.to_csv(os.path.join(data_path, "combined_data1.csv"), index=False)
```

```python
from markdownify import MarkdownConverter

def md(soup, **options):
    return MarkdownConverter(**options).convert(soup)
```

```python
test = md(combined_df['Nội dung'][0])
print(test)
```

```python
def  convert_to_md(text):
    if pd.isna(text):
        return text
    try:
        return  md(text)
    except:
        return text
combined_df['Nội dung'] = combined_df['Nội dung'].apply(lambda x: convert_to_md(x))
combined_df.to_csv(os.path.join(data_path, "combined_data2.csv"), index=False)
combined_df
combined_df.shape
```
