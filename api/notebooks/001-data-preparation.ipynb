````python
# write a function to extract entities from a text
def extract_entities(text, labels):
    entities = model.predict_entities(text, labels, threshold=0.7)
    metadata = {}
    for entity in entities:
        metadata[entity['label']] = list(set(metadata.get(entity['label'], []) + [entity['text']]))
    return metadata
````

````python
# contextual retriever
CONTEXT_PROMPT = """
<document>
{doc_content}
</document>

Here is the chunk we want to situate within the whole document:
<chunk>
{chunk_content}
</chunk>

Please give a short succinct context to situate this chunk within the overall document for the purposes of improving search retrieval of the chunk.
Answer only with the succinct context and nothing else. write in Vietnamese /no_think
"""

from ollama import chat

def situate_context(doc: str, chunk: str) -> str:
    response = chat(
        model="qwen3:1.7b",
        messages=[
            {
                "role": "user",
                "content": CONTEXT_PROMPT.format(doc_content=doc, chunk_content=chunk)
            }
        ]
    )
    return response['message']['content']
````

````python
from sqlalchemy import create_engine
import json

# Create MySQL connection
engine = create_engine('mysql+mysqlconnector://root:voithan@localhost:3306/bc_app')

# Transform metadata to match SkincarePlaceMetadata schema
def transform_metadata(metadata: str):
    try:
        # Handle already transformed metadata
        metadata_dict = json.loads(metadata)
        if any(key in metadata_dict for key in ['doctors', 'services', 'technologies', 'diseases', 'prices']):
            return metadata
    except:
        try:
            metadata_dict = ast.literal_eval(metadata)
        except:
            return '{}'  # Return empty JSON if parsing fails

    result = {}
    for key, value in metadata_dict.items():
        if key == 'Bác sĩ':
            result['doctors'] = value
        elif key == 'Dịch vụ':
            result['services'] = value
        elif key == 'Công nghệ':
            result['technologies'] = value
        elif key == 'Bệnh':
            result['diseases'] = value
        elif key == 'Giá cả':
            result['prices'] = value
    return json.dumps(result, ensure_ascii=False)  # Convert dict to JSON string

# Convert metadata column to standardized format
df['metadata'] = df['metadata'].apply(transform_metadata)

# Rename columns to match database schema
df_to_sql = df.rename(columns={
    'id': 'id',
    'Tên': 'name',
    'Nội dung': 'content', 
    'Quận/huyện': 'district',
    'Tỉnh/thành phố': 'city',
    'Nguồn': 'sourceArticle',
    'created_at': 'created_at',
    'updated_at': 'updated_at'
})

# Select only needed columns in correct order
df_to_sql = df_to_sql[[
    'id',
    'name',
    'content',
    'district', 
    'city',
    'sourceArticle',
    'metadata',
    'summary',
    'created_at',
    'updated_at'
]]

# Fill null values with empty strings for non-nullable columns
df_to_sql['district'] = df_to_sql['district'].fillna('')
df_to_sql['city'] = df_to_sql['city'].fillna('')
df_to_sql['sourceArticle'] = df_to_sql['sourceArticle'].fillna('')
df_to_sql['name'] = df_to_sql['name'].fillna('')

# Write to MySQL
df_to_sql.to_sql(
    'skincare_place',
    engine,
    if_exists='replace',
    index=False
)
````

````python
from markdownify import MarkdownConverter
from bs4 import BeautifulSoup
import re

def md(soup, **options):
    # Convert multiple <strong> tags into single <strong>
    if isinstance(soup, str):
        soup = BeautifulSoup(soup, 'html.parser')
    
    # Find all strong tags
    strong_tags = soup.find_all('strong')
    for tag in strong_tags:
        # Get all text content, including nested tags
        text = tag.get_text()
        # Create new strong tag with just the text
        new_tag = soup.new_tag('strong')
        new_tag.string = text
        # Replace old tag with new one
        tag.replace_with(new_tag)
        
    return MarkdownConverter(**options).convert(str(soup))
````
